```python
import sys
from collections import Counter

def CoinChangeRecursive(n,coins):
    q=sys.maxsize
    if n==0:
        return 0
    elif n>0:
        for i in coins:
            # print(n)
            q=min(q,1+CoinChangeRecursive(n-i,coins))
    return q
            

def BottomUpCoinChange(n,coins):
    r=[sys.maxsize]*(n+1)
    r[0]=0
    s=[0]*(n+1)
    for i in range(1,n+1): #Cada pos en la tabla
        for j in coins:
            if j<=n: 
                # print('i-j={}     r[i-j]={}'.format(i-j,r[i-j]))
                q=r[i-j] #subproblema 
                if 1+q<r[i]:
                    r[i]=1+q
                    s[i]=j  

    return r[n],s

def TopDownCoinChange(n,coins):
    r=[sys.maxsize]*(n+1)
    s=[0]*(n+1)
    return TopDownCoinChangeAux(n,coins,r,s)

def TopDownCoinChangeAux(n,coins,r,s):
    # print(n)
    q=0
    count=0
    if r[n]>=0 and r[n]!=sys.maxsize:#Ya esta solucionado
        return r[n],s[n]
    if n==0:
        q=0
    elif n>=0:
        q=sys.maxsize
        for i in coins:
            if i<=n:
                tmp=1+TopDownCoinChangeAux(n-i,coins,r,s)[0]
                if tmp<q:
                    q=tmp
                    s[n]=i
    r[n]=q

    return r[n],s


def printCoinChangeV2(n,coins,BottomUp=True):
    if BottomUp:          
        r,s=BottomUpCoinChange(n,coins)
    else:
        r,s=TopDownCoinChange(n,coins)  
    s_=list()
    while n>0:
        s_.append(s[n])
        n=n-s[n]
    tmp=dict(Counter(s_))
    tmp_={coin:0 for coin in coins}
    dict_coins={**tmp_, **tmp}
    return r,s_,dict_coins

coins= [1, 3, 21, 52]
n_arr= [0,1,3,6,9,11,23,24,29,62,69,75,
        81,95,103,154,125,250,310,1500,
        29001,378884,631453,2000000,2450254
        ,3000100,5499334,6453145,11000000]

coins_used=list()
for n in n_arr:
    r_,s_,coins_=printCoinChangeV2(n,coins,True)
    coins_used.append(coins_)
    print('n = {}  m = {}\nSolución: {}\nMonedas usadas: {}'.format(n,r_,s_,coins_))
```
